```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.font_manager as fm

# デフォルトフォント設定
font_path = "/System/Library/Fonts/ヒラギノ角ゴシック W6.ttc"
font_prop = fm.FontProperties(fname=font_path)
rcParams['font.family'] = font_prop.get_name()
```

```python
#特徴量の10個の場合

import matplotlib.pyplot as plt
import pandas as pd

# データ量と勝率
data = {'X日分の株価データ': ['1','3','9','16'],
        '勝率': [ 56.9, 49.1, 50.5, 52.0 ]}

# DataFrameに変換
df = pd.DataFrame(data)

# グラフ全体のフォント設定
plt.rc('font', size=12)  # デフォルトフォントサイズ
plt.rc('axes', titlesize=28, labelsize=28)  # タイトルとラベルサイズ
plt.rc('xtick', labelsize=28)  # X軸目盛りの文字サイズ
plt.rc('ytick', labelsize=28)  # Y軸目盛りの文字サイズ

# グラフ作成
plt.figure(figsize=(10, 8))
plt.plot(df['X日分の株価データ'], df['勝率'], color='b', marker='o', linestyle='-', label='勝率 (%)')

# ラベルやタイトルの設定
plt.xlabel('X日分の株価データ(日)')
plt.ylabel('勝率 (%)')
plt.title('特徴量10個の勝率予測')

# 目盛りや凡例の調整
plt.grid(True)
plt.ylim(45, 65)  # 勝率の範囲に合わせて調整
plt.legend(fontsize=28)  # 凡例を追加してフォントサイズを調整
plt.tight_layout()
plt.show()
```
