```python
import matplotlib.pyplot as plt
import pandas as pd

# サンプルデータ
data = {
    'Date': ['11月6日' , 
             '11月13日', 
             '11月20日',
             '11月27日',
             '12月3日' , 
             '12月11日', 
             '12月18日', 
             '12月25日', 
             '1月8日', 
             '1月15日', 
             '1月22日', 
            ],
    
    '勝率': [53.6, 54.7, 53.9, 54.7, 58.4, 58.2, 58.8, 56.8, 59.1, 57.4, 58.8]
}

# コメントはコード内にコメントアウトで記録
# 2024-10-30: 基本モデルでのテスト。改善の余地あり。

# 2024-11-06: 全て1mで設定。1d,2d,3d,4d,5dの順で記録、4本値データを標準化→特徴量の追加。外れ値の計算をしていない。　
#何度やっても、同じ値が出る→機械学習は、AIではないことがわかる。数学。

# 2024-11-13: 時系列クロスバリデーションの実行をした。　5個にデータを分けて、階段上になるやつ。
# 2024-11-20: ハイパラメータの使用。特徴量を自分で考えて、モデルにくらえることができる
# 2024-11-27: 


# DataFrameに変換
df = pd.DataFrame(data)


# グラフ全体のフォント設定
plt.rc('font', size=15)  # デフォルトフォントサイズ
plt.rc('axes', titlesize=28, labelsize=28)  # タイトルとラベルサイズ
plt.rc('xtick', labelsize=18)  # X軸目盛りの文字サイズ
plt.rc('ytick', labelsize=26)  # Y軸目盛りの文字サイズ

# グラフ作成
plt.figure(figsize=(10, 8))
plt.plot(df['Date'], df['勝率'], color='b', marker='o', linestyle='-', label='勝率 (%)')

# 各データポイントに数値を表示
for i, txt in enumerate(df['勝率']):
    plt.text(df['Date'][i], df['勝率'][i] + 0.3, f'{txt}', ha='center', fontsize=20, color='black')

# ラベルやタイトルの設定
plt.xlabel('予測した日')
plt.ylabel('勝率 (%)')
plt.title('勝率の予測結果')

# 目盛りや凡例の調整
plt.xticks(rotation=45)
plt.grid(True)
plt.ylim(50, 64)  # 勝率の範囲に合わせて調整
plt.legend(fontsize=28)  # 凡例を追加してフォントサイズを調整
plt.tight_layout()
plt.show()
```
