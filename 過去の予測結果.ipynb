```python
import matplotlib.pyplot as plt
import pandas as pd

# サンプルデータ
data = {
    'Date': ['2024-10-30', '2024-11-06', '2024-11-06', '2024-11-06', '2024-11-06',
             '2024-11-13', '2024-11-13', '2024-11-13', '2024-11-13',
             '2024-11-20', '2024-11-20', '2024-11-20', '2024-11-20'],
    'Hit Rate': [47.2, 53.6, 49.1, 51.7, 52.5, 54.7, 50.1, 49.2, 52.2, 57.9, 53.5, 52.5, 50.6]
}

# DataFrameに変換
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# 日付ごとの平均を計算
mean_values = df.groupby('Date')['Hit Rate'].mean().reset_index()

# グラフ作成
plt.figure(figsize=(10, 8))
plt.scatter(df['Date'], df['Hit Rate'], color='b', label='Hit Rate')  # 散布図
plt.plot(mean_values['Date'], mean_values['Hit Rate'], color='r', linestyle='-', label='Mean Hit Rate')  # 平均値の線

# ラベルやタイトルの設定
plt.xlabel('Date', fontsize=14)
plt.ylabel('Hit Rate (%)', fontsize=14)
plt.title('Prediction Result with Mean Line', fontsize=16)

# 目盛りや凡例を調整
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.ylim(45, 60)  # 勝率の範囲に合わせて調整
plt.legend(fontsize=12)  # 凡例の追加
plt.tight_layout()

plt.show()
```
