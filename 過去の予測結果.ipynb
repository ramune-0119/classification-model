```python
import matplotlib.pyplot as plt
import pandas as pd

# サンプルデータの初期設定（10月30日から1週間ごとに追加していく形にします）
data = {
    'Date': ['2024-10-30', '2024-11-06','2024-11-06','2024-11-06','2024-11-06','2024-11-06'],
    'Hit Rate': [47.2, 53.6,49.1,51.7,52.5,54.7]  
}

# コメントはコード内にコメントアウトで記録
# 2024-10-30: 基本モデルでのテスト。改善の余地あり。

# 2024-11-06: 全て1mで設定。1d,2d,3d,4d,5dの順で記録、4本値データを標準化→特徴量の追加。外れ値の計算をしていない。　
#何度やっても、同じ値が出る→機械学習は、AIではないことがわかる。数学。

# 2024-11-13: 
# 2024-11-20: 
# 2024-11-27: 

# DataFrameに変換
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# 散布図の作成
plt.figure(figsize=(10, 8))
plt.scatter(df['Date'], df['Hit Rate'], color='b')
plt.xlabel('Date')
plt.ylabel('Hit Rate (%)')
plt.title('Prediction result')
plt.grid(True)
plt.ylim(45, 60)  # 勝率の範囲に合わせて調整

# 各日程のデータポイント表示
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
